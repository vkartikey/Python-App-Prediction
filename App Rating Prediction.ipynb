# 8. Data Preprocessing

```python
inp1 = google_df.copy() #creating a copy of google_df
```

```python
inp1.head(10)
```

**8.1 Applying log transformation to 'Reviews' and 'Installs' Columns**

```python
inp1.skew() #To check skewness of the data
```

```python
inp1.describe()
```

```python
reviews_skew = np.log1p(inp1['Reviews']) #log transformation on Reviews column
inp1['Reviews'] = reviews_skew
```

```python
reviews_skew.skew()
```

```python
installs_skew = np.log1p(inp1['Installs']) #log transgormation on Installs column
inp1['Installs'] = installs_skew
```

```python
installs_skew.skew()
```

```python
inp1.head(10)
```

**8.2 Dropping Non-Usefull columns**

```python
inp1.drop(['App', 'Last Updated', 'Current Ver', 'Android Ver', 'Type'], axis = 1, inplace = True) #for dropping specified columns
```

```python
inp1.head(10)
```

```python
inp1.shape
```

**8.3 Dummy Encoding to convert Categorical variables to Numeric ones**

```python
inp2 = inp1 #creating a copy of data
```

```python
inp2.head(10)
```

**Applying Dummy encoding to Category Column**

**Category column**

```python
inp2['Category'].unique() #for knowing the unique values in the category column
```

```python
inp2.Category = pd.Categorical(inp2.Category)

x = inp2[['Category']]
del inp2['Category']

dummies = pd.get_dummies(x, prefix = 'Category')
inp2 = pd.concat([inp2, dummies], axis = 1)
inp2.head(10)
```

**Genres Column**

```python
inp2['Genres'].unique()
```

We are going to streamline the genres since there are too many genres and will assigning them to a single category called 'Other'

```python
lists = [] #Empty list

for i in inp2.Genres.value_counts().index: #getting total no. of genres and if less than 20 then assigning them to a new category Other
    if inp2.Genres.value_counts()[i] < 20:        lists.append(i)
        
inp2.Genres = ['Other' if i in lists else i for i in inp2.Genres] #changing the genres to other
```

```python
inp2['Genres'].unique()
```

```python
inp2.Genres = pd.Categorical(inp2['Genres']) #genres column into x variable
x = inp2[['Genres']]
del inp2['Genres'] #deleting the Genres column

#encoding and storing the dummies values in dataset
dummies = pd.get_dummies(x, prefix = 'Genres')
inp2 = pd.concat([inp2, dummies], axis = 1)
```

```python
inp2.head(10)
```

**Content Rating Column**

```python
inp2['Content Rating'].unique()
```

```python
inp2['Content Rating'] = pd.Categorical(inp2['Content Rating'])

x = inp2[['Content Rating']]
del inp2['Content Rating']

dummies = pd.get_dummies(x, prefix = 'Content Rating')
inp2 = pd.concat([inp2, dummies], axis = 1)
inp2.head(10)
```

# 9. Train and Test Splitting the Data
# 10. Separating the df into "X_train, X_test, y_train, y_test"

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error as mse
from sklearn import metrics
```

Creating X and Y variables; X = Independent Features; Y = Target Feature 

```python
ml1 = inp2
X = ml1.drop('Rating', axis = 1)
y = ml1['Rating']

X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.3, random_state = 5)
```

```python
print('Shape of X_train =', X_train.shape)
print('Shape of X_test =', X_test.shape)
print('Shape of y_train =', y_train.shape)
print('Shape of y_test =', y_test.shape)
```

# 11. Model Building

**11.1 Linear Regression Model Used**

```python
#Using Linear Regression technique
lr = LR()
lr.fit(X_train, y_train) #lr = LinearRegression
```

```python
print("Intercept of Linear Regression Model = ", lr.intercept_)
```

```python
print("Column Names for Co-Efficient =", ml1.columns)
print("Co-Efficient of Linear Regression Model = ", lr.coef_)
```

**11.2 R2 on Train Dataset**

```python
R2score_train = round(lr.score(X_train, y_train), 3)

print("R2 Score for the Training Data Set is", R2score_train)
```

# 12. Predictions on Test Set

```python
y_test_pred = lr.predict(X_test)

pd.DataFrame({'Actual y_test' : y_test,
             'Predict y_test' : y_test_pred}).head(25)
```

```python
R2score_test = round(lr.score(X_test, y_test), 3)

print("R2 Score for the Test Data Set is", R2score_test)
```
